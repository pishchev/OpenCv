```python
import cv2
import numpy as np
import random
```

```python
image = cv2.imread("1.jpg")

img_edge = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

param1 = 15
param2 = 15
dt = 1
mindist = 10
minrad = 10
maxrad = 100

cv2.namedWindow("trackBars")
def show():
    img = image.copy()

    circles = cv2.HoughCircles(img_edge, cv2.HOUGH_GRADIENT, dt, mindist, param1=param1, param2=param2, minRadius=minrad, maxRadius=maxrad)
    circles = np.uint16(np.around(circles))
    for i in circles[0,:]:
        # нарисовать окружности
        cv2.circle(img,(i[0],i[1]),i[2],(0,255,0),2)
        # нарисовать центры окружностей
        cv2.circle(img,(i[0],i[1]),2,(0,0,255),3)
    cv2.imshow("pic",img)

show()

def p1(x):
    global param1
    param1 = x
    show()

def p2(x):
    global param2
    param2 = x
    show()

def dt1(x):
    global dt
    dt = x
    show()
    
def mind(x):
    global mindist
    mindist = x
    show()
    
def minr(x):
    global minrad
    minrad = x
    show()

def maxr(x):
    global maxrad
    maxrad = x
    show()

cv2.createTrackbar("param1", "trackBars",0,255, p1)
cv2.createTrackbar("param2","trackBars" ,0,255, p2)
cv2.createTrackbar("dt", "trackBars",0,255, dt1)
cv2.createTrackbar("mindist", "trackBars",0,255, mind)
cv2.createTrackbar("minrad", "trackBars",0,255, minr)
cv2.createTrackbar("maxrad", "trackBars",0,255, maxr)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
